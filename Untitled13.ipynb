```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen= ImageDataGenerator(
    rescale= 1./255,
    shear_range= 0.2,
    zoom_range= 0.2,
    horizontal_flip= True,
    vertical_flip= True,
    rotation_range= 40,
    width_shift_range= 0.2,
    height_shift_range= 0.2,
    brightness_range= [0.8, 1.2],
    fill_mode= 'nearest'
)

test_datagen = ImageDataGenerator(rescale= 1./255)

train_datagen = train_datagen.flow_from_directory(
    '/content/Teeth_Dataset/Training',
    target_size=(256, 256),
    batch_size=32,
    class_mode='categorical'
)

validation_generator = test_datagen.flow_from_directory(
    '/content/Teeth_Dataset/Validation',
    target_size=(256, 256),
    batch_size=32,
    class_mode='categorical'
)
```

```python
from collections import OrderedDict
# Retrieve the class labels from the training data generator
cls_train = train_datagen.classes

# Retrieve the class names from the training data generator
classes = list(train_datagen.class_indices.keys())

# Compute the number of samples per class
unique, counts = np.unique(cls_train, return_counts=True)
valdict = OrderedDict(zip(unique, counts))

# Create a list to store the number of samples per class
num_values = []
for i in range(len(classes)):
    num_values.append(valdict[i])

# Plot the number of samples per class
plt.figure(figsize=(30, 10))
x = np.arange(len(classes))
plt.bar(x, num_values)
plt.xlabel('Classes')
plt.ylabel('Number of Samples')
plt.title('Number of Samples per Class')
plt.xticks(x, classes)
plt.show()
```

```python

# Retrieve the class labels from the validation (test) data generator
cls_test = validation_generator.classes

# Retrieve the class names from the validation data generator
classes = list(validation_generator.class_indices.keys())

# Compute the number of samples per class
unique, counts = np.unique(cls_test, return_counts=True)
valdict = OrderedDict(zip(unique, counts))

# Create a list to store the number of samples per class
num_values = []
for i in unique:
    num_values.append(valdict[i])

# Plot the number of samples per class
plt.figure(figsize=(30, 10))
x = np.arange(len(classes))
plt.bar(x, num_values)
plt.xlabel('Classes')
plt.ylabel('Number of Samples')
plt.title('Number of Samples per Class in Test Data')
plt.xticks(x, classes)
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, BatchNormalization, ReLU, Add, GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras import Model, Input

input_shape = (256, 256, 3)
inputs = Input(shape=input_shape)

x = Conv2D(64, (7, 7), strides=(2, 2), padding='same', use_bias=False)(inputs)
x = BatchNormalization()(x)
x = ReLU()(x)
x = tf.keras.layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2), padding='same')(x)

residual = x
x = Conv2D(64, (3, 3), padding='same', use_bias=False)(x)
x = BatchNormalization()(x)
x = ReLU()(x)
x = Conv2D(64, (3, 3), padding='same', use_bias=False)(x)
x = BatchNormalization()(x)
x = Add()([x, residual])
x = ReLU()(x)

residual = Conv2D(128, (1, 1), strides=(2, 2), padding='same', use_bias=False)(x)
x = Conv2D(128, (3, 3), strides=(2, 2), padding='same', use_bias=False)(x)
x = BatchNormalization()(x)
x = ReLU()(x)
x = Conv2D(128, (3, 3), padding='same', use_bias=False)(x)
x = BatchNormalization()(x)
x = Add()([x, residual])
x = ReLU()(x)

residual = Conv2D(256, (1, 1), strides=(2, 2), padding='same', use_bias=False)(x)
x = Conv2D(256, (3, 3), strides=(2, 2), padding='same', use_bias=False)(x)
x = BatchNormalization()(x)
x = ReLU()(x)
x = Conv2D(256, (3, 3), padding='same', use_bias=False)(x)
x = BatchNormalization()(x)
x = Add()([x, residual])
x = ReLU()(x)

residual = Conv2D(512, (1, 1), strides=(2, 2), padding='same', use_bias=False)(x)
x = Conv2D(512, (3, 3), strides=(2, 2), padding='same', use_bias=False)(x)
x = BatchNormalization()(x)
x = ReLU()(x)
x = Conv2D(512, (3, 3), padding='same', use_bias=False)(x)
x = BatchNormalization()(x)
x = Add()([x, residual])
x = ReLU()(x)

x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
outputs = Dense(7, activation='softmax')(x)

Teeth_cnn = Model(inputs=inputs, outputs=outputs)

Teeth_cnn.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

Teeth_cnn.summary()

# Fixing the training process
batch_size = 32

train_steps = len(train_datagen) // batch_size
validation_steps = len(validation_generator) // batch_size

# Ensure that train_datagen and validation_generator are defined and returning correct data
train_steps = max(train_steps, 1)  # Ensure at least one step per epoch
validation_steps = max(validation_steps, 1)  # Ensure at least one step per epoch

model_history = Teeth_cnn.fit(
    train_datagen,
    epochs=70,
    steps_per_epoch=train_steps,
    validation_data=validation_generator,
    validation_steps=validation_steps
)
```
